# Weather Prediction Using Naive Bayes

## Bayes Formula

P(A/B) = P(B/A)*P(A)P(B)

P(A) = Probability of event A

P(B) = Probability of event B

P(A/B) = Probability of event A given event B has already occured

P(B/A) = Probability of event B given event A has already occures

```python
# import libraries

import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
# download the dataset

!wget https://raw.githubusercontent.com/gr8Adakron/naive-bayes-using-python/master/tennis.csv
```

```python
# read csv file

df = pd.read_csv('tennis.csv')
df
```

```python
# encoding the strings to numericals

outlook_at = LabelEncoder()
Temp_at = LabelEncoder()
Hum_at = LabelEncoder()
win_at = LabelEncoder()
numerics = LabelEncoder()
```

```python
# dropping the target variable and make it as newframe

inputs = df.drop('Play', axis='columns')
target = df['Play']
target
```

```python
# creating the new dataframe

inputs['outlook_n'] = outlook_at.fit_transform(inputs['Outlook'])
inputs['Temp_n'] = Temp_at.fit_transform(inputs['Temp.'])
inputs['Hum_n'] = Hum_at.fit_transform(inputs['Humidity'])
inputs['win_n'] = win_at.fit_transform(inputs['Windy'])
inputs
```

```python
# dropping the string values

inputs_n = inputs.drop(['Outlook', 'Temp.', 'Humidity', 'Windy'], axis='columns')
inputs_n
```

```python
# applying the gaussian naivebayes

classifier = GaussianNB()
classifier.fit(inputs_n, target)
```

```python
# accuracy

classifier.score(inputs_n, target)
```

```python
# prediction

classifier.predict([[1, 0, 0, 1]])
```
